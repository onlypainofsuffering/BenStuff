```python
def lingkaran(r):
    luas = 3.14 * r * r
    keliling = 2 * 3.14 * r
    return luas, keliling

luas, keliling = lingkaran(7)
print("Luas:", luas)
print("Keliling:", keliling)
```

```python
usia = int(input("Masukkan usia: "))

if usia >= 0 and usia <= 13:
    print("Anak")
elif usia >= 14 and usia <= 24:
    print("Remaja")
elif usia >= 25 and usia <= 49:
    print("Dewasa")
else:
    print("Lansia")
```
